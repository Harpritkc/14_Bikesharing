```python
import pandas as pd
import os
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
nyc_biketrip_df = pd.read_csv("201908-citibike-tripdata.csv", low_memory=False)
nyc_biketrip_df.head()
```

```python
# 2. Check the datatypes of your columns. 
nyc_biketrip_df.info()
```

```python
# Check to see how many trips are greater than hour.
nyc_biketrip_over_hr = nyc_biketrip_df.loc[nyc_biketrip_df['tripduration']>3600]
nyc_biketrip_over_hr
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
nyc_biketrip_df['tripduration'] = pd.to_datetime(nyc_biketrip_df['tripduration'], unit='s')
nyc_biketrip_df.head() 
```

```python
# 4. Check the datatypes of your columns. 
nyc_biketrip_df.info()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
nyc_biketrip_df.to_csv("201908-citibike-tripdata_2.csv", index=False)
```
